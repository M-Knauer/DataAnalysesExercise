```python
import csv
from datetime import datetime as dt
import requests as r
```

```python
url = "https://api.covid19api.com/country/brazil"

req = r.get(url)

if req.status_code == 200:
    print("OK!")
else:
    print("ERROR!")
```

```python
dados = req.json()
```

```python
final_data = [["Confirmados", "Mortes", "Recuperados", "Ativos", "Data"]]

for v in dados:
    final_data.append([v["Confirmed"], v["Deaths"], v["Recovered"], v["Active"], v["Date"][:10]])
final_data
```

```python
CONFIRMADOS = 0
MORTES = 1
RECUPERADOS = 2
ATIVOS = 3
DATA = 4
```

```python
with open("brasil_covid.csv", "w", newline="", encoding="UTF-8") as file:
    writer = csv.writer(file)
    writer.writerows(final_data)
```

```python
for i in range(1, len(final_data)):
    final_data[i][DATA] = dt.strptime(final_data[i][DATA], "%Y-%m-%d").date()
```

```python
def get_dataset(y, labels):
    if type(y) == list:
        dataset = []
        for i in range(len(y)):
            dataset.append({
                "label": labels[i],
                "data": y[i]
            })
        return dataset
    else:
        return [{
            "label": labels[0],
            "data": y
        }]
```

```python
def set_title(title=""):
    if title != "":
        display = "true"
    else:
        display = "false"
    
    return {
        "title": title,
        "display": display
    }
```

```python
def create_chart(x, y, labels, kind="bar", title=""):
    
    datasets = get_dataset(y, labels)
    options = set_title(title)
    
    chart = {
        "type": kind,
        "data": {
            "labels": x,
            "datasets": datasets
        },
        "options": options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = "https://quickchart.io/chart"
    req = r.get(f"{url_base}?c={str(chart)}")
    return req.content
```

```python
def save_image(path, content):
    with open(path, "wb") as image:
        image.write(content)
```

```python
y_data1 = []
for v in final_data[1::10]:
    y_data1.append(v[CONFIRMADOS])

y_data2 = []
for v in final_data[1::10]:
    y_data2.append(v[RECUPERADOS])

labels = ["CONFIRMADOS", "RECUPERADOS"]

x = []
for v in final_data[1::10]:
    x.append(v[DATA].strftime("%d/%m/%Y"))

chart = create_chart(x, [y_data1, y_data2], labels, title="Grfico confirmados e recuperados")
chart_content = get_api_chart(chart)
save_image("meu-primeiro-grafico.png", chart_content)
display_image("meu-primeiro-grafico.png")
    
```
